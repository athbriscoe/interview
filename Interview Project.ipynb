```python
import numpy as np

#this calculates the dihedral angle between 4 specified coordinates.
#atomic coordinates are specified and entered into numpy array form.
a = np.array([-0.88, 1.23, -0.82]) 
b = np.array([0.00, 1.17, -0.19])
c = np.array([0.00, 0.00, 0.57])
d = np.array([0.00, -1.17, -0.19])

#Calcluating vectors from the atomic coordinates
v0 = -1*(b-a)
v1 = c-b  
v2 = d-c  

#cross products for the vectors.
xp1 = np.cross(v0, v1)
xp2 = np.cross(v2, v1)

fullxp = np.cross(xp1, xp2)

#calculate dot products for x and y values.
#vector 1 has been normalised to account during the y calculation.
y = np.dot(fullxp, v1)*(1/np.linalg.norm(v1))
x = np.dot(xp1, xp2)

#angle is required in degrees.
angle = np.degrees(np.arctan2(y, x))
print(angle)
```

```python
%%time
#import Psi4 as a Python module.
#%%time for calculation time tracking, but actually not needed here due to psi4 capabilities.
import psi4 

#specify bond lengths (R, Angstroms), simple angles (A, degrees) and torsional/dihedral
#angles (D, degrees).
R1 = 1.38
R2 = 1.07
A1 = 120.00
D1 = 0.00
D2 = 180.00

#construct Z-matrix for benzene using geometry from reference energy value.
#this allows for easier comparison.
#Z-matrix has been constructed via string formatting for easier editing if required.
benzene = psi4.geometry("""
C
C 1 {0}
C 2 {0} 1 {1} 
C 3 {0} 2 {1} 1 {2}
C 4 {0} 3 {1} 2 {2}
C 5 {0} 4 {1} 3 {2}
H 1 {3} 2 {1} 3 {4}
H 2 {3} 3 {1} 4 {4}
H 3 {3} 4 {1} 5 {4}
H 4 {3} 5 {1} 6 {4}
H 5 {3} 6 {1} 1 {4}
H 6 {3} 1 {1} 2 {4}
""". format(R1, A1, D1, R2, D2)
)

#Calculate the SCF energy for each basis set defined using a 'for loop'.
basis_sets = ['hf/3-21g', 'hf/6-31g', 'hf/6-31g**', 'hf/cc-pVDZ', 'hf/cc-pVTZ', 'hf/cc-pVQZ',
              'hf/cc-pV5Z',
              'PBE/3-21g', 'PBE/6-31g', 'PBE/6-31g**', 'PBE/cc-pVDZ', 'PBE/cc-pVTZ']
for basis_set in basis_sets:
    psi4.energy(basis_set, molecule=benzene)
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Reference value for ground state benzene energy taken from CCCBDB. 
#Uses CCSD(T)=full/aug-cc-pVTZ.
#values have been rounded to 3 decimal places for efficient plotting and comparison.
benzene_ref = -231.936
benzene_hf_values = np.array([-229.419, -230.624, -230.713, -230.721, -230.780, -230.795, -230.798])

#calculating the absolute percentage error from reference for each model basis set.
new_values = [benzene_hf_value - benzene_ref for benzene_hf_value in benzene_hf_values]
hf_accuracy = np.abs([new_value*100/benzene_ref for new_value in new_values])

#times are taken from 'system module time' in psi4 outputs.
benzene_hf_times = np.array([3, 2, 9, 4, 17, 30, 86])

benzene_pbe_values = np.array([-230.658, -231.886, -231.939, -231.944, -232.011])
new_values = [benzene_pbe_value - benzene_ref for benzene_pbe_value in benzene_pbe_values]
pbe_accuracy = np.abs([new_value*100/benzene_ref for new_value in new_values])
benzene_pbe_times = np.array([10, 4, 14, 16, 29])

#plotting the 2 data sets.
plt.scatter(
    x=hf_accuracy,
    y=benzene_hf_times
)
plt.scatter(
    x=pbe_accuracy,
    y=benzene_pbe_times
    )

hf_labels = ["3-21G", "6-31G", "6-31G**", "cc-pVDZ", "cc-pVTZ", "cc-pVTQZ", "cc-pV5Z"]
pbe_labels = ["3-21G", "6-31G", "6-31G**", "cc-pVDZ", "cc-pVTZ"]
plt.xlabel("Absolute Percentage Error")
plt.ylabel("Time / sec, 10^-2")
plt.title("Error vs Time Taken for Benzene Calculations")
plt.legend(["Hartree Fock Basis Set", "PBE DFT Basis Set"])

#adding labels to the scatter plot for easier identification of each data point.
for i, label in enumerate(hf_labels):
    plt.text(hf_accuracy[i], benzene_hf_times[i]-0.5, label, ha='right', rotation=-5)
for i, label in enumerate(pbe_labels):
    plt.text(pbe_accuracy[i-2], benzene_pbe_times[i-2]+0.2, label)
    
plt.show()
```
